```python
d=data.head(10).reset_index()
```

```python
plt.plot(d["pH"],label="pH",linewidth=4.)
plt.plot(d["quality"],label="Quality",color="yellow",linestyle="-.")
plt.title("PH and Quality of red wine",fontsize=14)
plt.legend(loc="upper left",fontsize=10)
plt.show
          
```

### The quality gradually and  constantly begins from 0 to 2 on the x-axis and experiences a rise from 5.0 to 6.0 on the y-axis.The quality then drops abruptly from 6.0   back to 5.0 on the y-axis and continues gradually from 4.0 to 6.0 on the x-axis.It experiences another rise from 5.0 to 7.0 on the y-axis and moves at a constant pace  and stops at 8.0 on the x-axis.Finally it drops from 7.0 to 5.0 on the y-axis.

### The pH experiences a decrease from 3.5 to 3.2 and then a rise from 3.2 to 3.26 .Again it experiences a decrease from 3.26 to 3.16 and then further increases to 3.51.A drop occurs from 3.51 to 3.45 and then an increase to 3.52 and then a further decrease to 3.42.Finally it increases to 3.57 and then decreases to 3.39.


```python
x=data.alcohol
y=data.pH
plt.scatter(x,y)
plt.title('A SCATTERED GRAPH OF ALCOHOL AGAINST pH')
plt.xlabel('Alcohol')
plt.ylabel('pH')
plt.show
```

### There appears to be a strong association between alcohol and pH .It is a non-linear graph and there are obvious outliers.

 

```python
x=data.sulphates
y=data.chlorides
plt.scatter(x,y)
plt.title('A SCATTERED GRAPH OF SULPHATES  AGAINST CHLORIDES')
plt.xlabel('Sulphates')
plt.ylabel('Chlorides')
plt.show
```

### There is a strong association between chlorides and sulphates from 0.25 to 1.25 0n the x-axis and 0.0 to 0.3 on the y-axis.The graph is non-linear and there are many outliers.
